### Introduction to Neural Network Regression in Tensorflow

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

**Custom Data**

```python
# Creating features
X = np.arange(-7,15,3,dtype=np.float32)

# Create labels 
y=np.arange(3,3*len(X)+1,3,dtype=np.float32)

# visualization:
plt.scatter(X,y)
plt.show()
```

**4) Improving the model**

We can improve our model, by altering the steps we took to create a model. (basically adjusting hyper-params)
* Creating the model:
  * Add more layers
  * Increase number of hidden units (neurons) within each of these layers
  * Change the activation function
* Compiling the model:
  * Change the optimization function or learning rate of the optimization function
  * Even loss function changes can help
* Fitting the model
  * Increase number of epochs

* Fitting on more training data

```python
# convert data in tensor:
X=tf.constant(X)
y=tf.constant(y)

# setting the random seed
tf.random.set_seed(42)

# 1. creating model using sequential API
model=tf.keras.Sequential([
    tf.keras.layers.Dense(1) # input shape: 1, output shape: 1
])


# 2. compile the model
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=["mae"])
```

**EXP 1: Training for longer epochs:**

```python
model.fit(tf.expand_dims(X,-1),y, epochs=100)
```

```python
# 4. predicting using model:
model.predict(tf.constant([17.0]))
```

**EXP 2: Adding a hidden layer with 10 hidden units**

```python
# 1. creating model using sequential API
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10,activation='relu')) # hidden layer with 10 units (relu activation)
model.add(tf.keras.layers.Dense(1)) # output layer with 1 unit (linear activation)

# 2. compile the model
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=["mae"])

# 3. fitting the model
model.fit(tf.expand_dims(X,-1),y, epochs=100)
```

```python
# 4. predicting using model:
model.predict(tf.constant([17.0]))
```

**EXP 3: Changing optimization function and Learning rate**

```python
# 1. creating model using sequential API
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10,activation='relu')) # hidden layer with 10 units (relu activation)
model.add(tf.keras.layers.Dense(1)) # output layer with 1 unit (linear activation)

# 2. compile the model
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),metrics=["mae"])

# 3. fitting the model
model.fit(tf.expand_dims(X,-1),y, epochs=100)
```

```python
# 4. predicting using model:
model.predict(tf.constant([17.0]))
```
